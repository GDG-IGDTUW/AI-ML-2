````python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
````

````python
bank_data = pd.read_csv('UniversalBank.csv')
````

````python
print(bank_data.head())
````

````python
print(bank_data.info()) #no null values
````

````python
print(bank_data.describe())
````


Countplot of PERSONAL LOAN
```
# This is formatted as code
```



````python
sns.countplot(x='Personal Loan', data=bank_data)
plt.xlabel("Personal Loan (0 = No, 1 = Yes)")
plt.ylabel("Count")
plt.show()
````

Countplot for EDUCATION

````python
sns.countplot(x='Education', data=bank_data)
plt.xlabel("Education Level (1 = Undergrad, 2 = Graduate, 3 = Advanced/Professional)")
plt.ylabel("Count")
plt.show()
````

Countplot for CREDIT CARD

````python
sns.countplot(x='CreditCard', data=bank_data)
plt.xlabel("Credit Card (0 = No, 1 = Yes")
plt.ylabel("Count")
plt.show()
````

Countplot for AGE

````python
bins = [0, 25, 35, 45, 55, 65, 100]
labels = ['0-25', '26-35', '36-45', '46-55', '56-65', '65+']
bank_data['Age Group'] = pd.cut(bank_data['Age'], bins=bins, labels=labels)

sns.countplot(x='Age Group', data=bank_data, order=labels)
plt.xlabel("Age Group")
plt.ylabel("Count")
plt.show()
````
